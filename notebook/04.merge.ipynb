```python
import pandas as pd
df = pd.read_csv("입지후보지.csv")

df1 = pd.read_csv("성환읍_후보지(MCLP).csv")
df2 = pd.read_csv("성환읍_후보지(P-median).csv")

# 공통 컬럼 기준 교집합 merge
merged = pd.merge(df1, df2, on="주소", how="inner")

# 중복 제거
merged1 = merged.drop_duplicates()
merged1.drop(['위도_y', '경도_y', '이름_y'], axis=1, inplace=True)
merged1['행정동'] = '성환읍'
merged1
```

```python
df1 = pd.read_csv("청룡동_후보지(MCLP).csv")
df2 = pd.read_csv("청룡동_후보지(P-median).csv")

# 공통 컬럼 기준 교집합 merge
merged = pd.merge(df1, df2, on="주소", how="inner")

# 중복 제거
merged2 = merged.drop_duplicates()
merged2.drop(['위도_y', '경도_y', '이름_y'], axis=1, inplace=True)
merged2['행정동'] = '청룡동'
merged2
```

```python
df1 = pd.read_csv("신안동_후보지(MCLP).csv")
df2 = pd.read_csv("신안동_후보지(P-median).csv")

# 공통 컬럼 기준 교집합 merge
merged = pd.merge(df1, df2, on="이름", how="inner")

# 중복 제거
merged3 = merged.drop_duplicates()
merged3.drop(['위도_y', '경도_y', '주소_y'], axis=1, inplace=True)
merged3.columns = ['이름_x', '위도_x', '경도_x', '주소']
merged3['행정동'] = '신안동'
merged3
```

```python
result = pd.concat([merged1, merged2, merged3]).reset_index(drop=True)
result.columns = ['이름', '위도', '경도', '주소', '행정동']
```

```python
result.to_csv("최종후보지.csv", index=False, encoding="utf-8-sig")
```
